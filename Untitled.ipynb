```python
# Удаляем строки с любыми NaN
df_cleaned = df.dropna()
print("DataFrame после удаления строк с NaN:")
print(df_cleaned)
```

```python
top5_by_count = df_cleaned['service'].value_counts().head(5)
print("Топ-5 услуг по количеству заказов:")
print(top5_by_count)
```

```python
# Группируем по городу и считаем среднее
avg_by_city = df_cleaned.groupby('city')['amount'].mean().reset_index()

# Переименовываем колонку для ясности

avg_by_city = avg_by_city.rename(columns={'amount': 'average_amount'})

print("Средняя сумма транзакций по городам:")
print(avg_by_city)
```

```python
# Сортируем от большего к меньшему
avg_by_city_sorted = avg_by_city.sort_values('average_amount', ascending=False)

print("\nТоп  10  городов по средней сумме транзакций:")
print(avg_by_city_sorted.head(10)) 
```

```python
# Определить услугу с наибольшей выручкой (по сумме amount). 
revenue_by_service = df_cleaned.groupby('service')['amount'].sum()

# заменаь ФОРМАТА  9.821188e+07 → 98,211,880
formatted_revenue = revenue_by_service.apply(lambda x: f"{x:,.0f}")

print("Сумма выручки по услугам:")
print(formatted_revenue)
```

```python
# Подсчет количества транзакций по каждому способу оплаты
payment_distribution = df_cleaned['payment_method'].value_counts()

#  процентного соотношения
payment_percent = df_cleaned['payment_method'].value_counts(normalize=True) * 100

# 
payment_stats = pd.DataFrame({
    'Количество транзакций': payment_distribution,
    'Процентное соотношение': payment_percent.round(2).astype(str) + '%'  # Добавляем знак процента
})

print("Распределение транзакций по способам оплаты:")
print(payment_stats)
```

```python
# загрузка данных 
clients_df = pd.read_json('clients_data.json')

#  Проверка структуры
print("Колонки в клиентских данных:", clients_df.columns.tolist())
print("Колонки в основном DataFrame:", df_cleaned.columns.tolist())
```

```python
#  3.	Объединение данных. Объединить транзакции с клиентскими данными (по id). 
# лпределили что в json файле id а в нашем data 'client_id'
merged_df = pd.merge(
    df_cleaned,
    clients_df,
    left_on='client_id',
    right_on='id',
    how='inner'  # Только где client_id == id
)
print(merged_df)
```

```python
merged_df = merged_df.drop(columns=['id']) # удоляем лищнею колонку про id 
print(merged_df)
```

```python
# Условия для категорий
conditions = [
    (merged_df['net_worth'] < 100000),
    (merged_df['net_worth'].between(100000, 1000000)),
    (merged_df['net_worth'] > 1000000)
]

# Названия категорий
choices = ['Низкий капитал', 'Средний капитал', 'Высокий капитал']

# Создание колонки
merged_df['asset_level'] = pd.cut(
    merged_df['net_worth'],
    bins=[-float('inf'), 100000, 1000000, float('inf')],
    labels=choices,
    right=False
)
print(merged_df)
```

```python
# Группировка по уровню активов
revenue_by_asset = merged_df.groupby('asset_level', observed=False)['amount'].agg(['sum', 'count', ]).apply(lambda x: x.apply("{:,.2f}".format))

# Переименование колонок
revenue_by_asset = revenue_by_asset.rename(columns={
    'sum': 'выручка',
    'count': 'Количество транзакций',
})

# Сортировка по выручке
revenue_by_asset = revenue_by_asset.sort_values('выручка', ascending=False)

print("Выручка по категориям клиентов:")
print(revenue_by_asset)
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Конвертируем дату в формат datetime и извлекаем месяц и год
merged_df['transaction_date'] = pd.to_datetime(merged_df['transaction_date'])
merged_df['month_year'] = merged_df['transaction_date'].dt.to_period('M')

# Агрегируем данные по месяцам (количество транзакций и сумма)
monthly_data = merged_df.groupby('month_year').agg(
    transactions_count=('transaction_id', 'count'),
    total_amount=('amount', 'sum')
).reset_index()

# Преобразуем период в timestamp для работы с числами
monthly_data['month_num'] = monthly_data['month_year'].dt.to_timestamp().astype('int64') // 10**9
```

```python
plt.figure(figsize=(12, 6))
plt.plot(monthly_data['month_year'].astype('str'), monthly_data['transactions_count'], marker='o')
plt.title('Динамика количества транзакций по месяцам', fontsize=15)
plt.xlabel('Месяц', fontsize=12)
plt.ylabel('Количество транзакций', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# После создания monthly_data добавим:
monthly_data = monthly_data.reset_index(drop=True)

# Подготовка данных для модели
X = monthly_data[['month_num']]  # Используем DataFrame с именем признака
y = monthly_data['transactions_count']

# Обучение модели
model = LinearRegression()
model.fit(X, y)  # Модель запомнит имена признаков

# Прогноз на следующий месяц
next_month_num = monthly_data['month_num'].iloc[-1] + 30*24*60*60  # Добавляем 1 месяц в секундах

# Важно: подаем на прогноз DataFrame с тем же именем признака
next_month_pred = model.predict(pd.DataFrame({'month_num': [next_month_num]}))

print(f"Прогнозируемое количество транзакций: {int(next_month_pred[0])}")
```

```python
# Создаем датуфрейм с прогнозом
future_month = monthly_data['month_year'].max() + 1
predictions = pd.DataFrame({
    'month_year': list(monthly_data['month_year']) + [future_month],
    'values': list(y) + [next_month_pred[0]],
    'type': ['actual']*len(y) + ['predicted']
})

plt.figure(figsize=(12, 6))
for typ, group in predictions.groupby('type'):
    plt.plot(group['month_year'].astype('str'), group['values'], 
             marker='o' if typ == 'actual' else 's',
             label='Фактические данные' if typ == 'actual' else 'Прогноз')
plt.title('Прогноз количества транзакций на следующий месяц', fontsize=15)
plt.xlabel('Месяц', fontsize=12)
plt.ylabel('Количество транзакций', fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Группируем данные по месяцам
monthly_demand = merged_df.groupby(pd.to_datetime(merged_df['transaction_date']).dt.to_period('M'))['amount'].sum()

# Прогноз как среднее последних 3 месяцев
last_3_months = monthly_demand.iloc[-3:].mean()
print(f"Прогноз на следующий месяц: {last_3_months:.2f}")
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Создаем временные признаки
X = np.array(range(len(monthly_demand))).reshape(-1, 1)  # Номера месяцев как признак
y = monthly_demand.values

# Обучаем модель
model = LinearRegression()
model.fit(X, y)

# Прогноз для следующего месяца
next_month = len(monthly_demand)
forecast = model.predict([[next_month]])[0]
print(f"Прогноз по линейной регрессии: {forecast:.2f}")

# Визуализация
plt.figure(figsize=(10,5))
plt.plot(monthly_demand.index.astype(str), y, label='Факт')
plt.plot([*monthly_demand.index.astype(str), 'Next'], 
         [*y, forecast], '--r', label='Прогноз')
plt.title('Прогноз спроса на следующий месяц')
plt.legend();
```
